```python
from tspqaoa import optimization, qls, utils
import networkx as nx
import random
```

```python
G = nx.complete_graph(10)
for (u,v,w) in G.edges(data=True):
    w['weight'] = random.random()

pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, with_labels=True)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
```

```python
G_qls = qls.graph_rewrite(G, '0123456789', '234')

pos = nx.spring_layout(G_qls)
nx.draw_networkx(G_qls, pos, with_labels=True)
labels = nx.get_edge_attributes(G_qls, 'weight')
nx.draw_networkx_edge_labels(G_qls, pos, edge_labels=labels)
```
